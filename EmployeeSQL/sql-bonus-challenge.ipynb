```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:!@localhost:5432/SQL-Challenge')
conn = engine.connect()
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('Salary')
for column in columns:
    print(column["name"], column["type"])
```

```python
employee = pd.read_sql('Select * from "Employee"', conn)
employee
```

```python
salary = pd.read_sql('Select * from "Salary"', conn)
salary
```

```python
titles = pd.read_sql('select * from "Titles"', conn)
titles
```

```python
salary['salary'] = salary['salary'].str.lstrip('$')

salary['salary'] = salary['salary'].str.replace(',','').astype(float)
```

```python
combined = pd.merge(salary, titles, on='emp_no', how='left')
combined
```

```python
title_salary = combined[['title', 'salary']]
```

```python
salary_type = combined.groupby("title").mean()["salary"].round(2)
salary_type
```

```python
#2. Create a histogram to visualize the most common salary ranges for employees.
title_salary.hist(column='salary')
plt.xlabel("Salary Range")
plt.ylabel("Salary Frequency");
```

```python
#3. Create a bar chart of average salary by title.
salary_type.plot.bar()
```
